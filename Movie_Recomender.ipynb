```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('datasets/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies = movies[['id','title', 'genres', 'keywords', 'overview','production_companies', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['production_companies'] = movies['production_companies'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['production_companies'][0]
```

```python
def extract_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 4:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(extract_cast)
```

```python
movies['cast'][0]
```

```python
def extract_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(extract_director)
```

```python
movies['crew'][1000]
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['production_companies'] + movies['cast'] + movies['crew']
```

```python
movies['tags']
```

```python
movies_df = movies[['id','title','tags']]
movies_df.head(1)
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:" ".join(x))
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(word):
    y = []
    for i in word.split():
        y.append(ps.stem(i))
    return " ".join(y)

movies_df['tags'] = movies_df['tags'].apply(stem)
movies_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cvectoriser = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cvectoriser.fit_transform(movies_df['tags']).toarray()
```

```python
print(vectors)
```

```python
cvectoriser.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matx = cosine_similarity(vectors)
```

```python
similarity_matx
```

```python
def recommend(movie):
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    distances = similarity_matx[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```

### Model Evaluation
Because we built a **Content-Based Filtering** system using Cosine Similarity, there is no traditional "Accuracy Score" (like 95% or 0.85) to calculate. For our movie recommender, we don't have user watch-history data to test if a user actually clicked on the recommended movies.

However, we can **qualitatively justify that our model is excellent** by looking at the results from the `recommend('Iron Man')` test:

1. **It understands Franchises:** All 5 recommendations belong to the exact same Marvel Cinematic Universe (MCU).
2. **Direct Sequels are Prioritized:** It correctly identified that the most mathematically similar movies to "Iron Man" are its direct sequels (Iron Man 2 & 3).
3. **Cross-over Recognition:** It recommended the "Avengers" and "Captain America: Civil War". Even though "Iron Man" isn't in the title of these movies, the model recognized that they share the exact same main cast (Robert Downey Jr.), share the same production companies (Marvel Studios), and have identical keywords (superhero, comic book).

This proves that our text-vectorization and cosine similarity approach worked quite well! The model accurately understands the context, genre, and connections between movies.
